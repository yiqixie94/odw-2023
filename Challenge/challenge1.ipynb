```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import gwosc
import gwpy
from gwpy.timeseries import TimeSeries
from pycbc.frame import read_frame
```

```python
ts = read_frame("challenge1.gwf", "H1:CHALLENGE1")
ts.plot()
```

```python
ts.sample_rate
```

```python
ts.duration
```

```python
ts_gwpy = TimeSeries.from_pycbc(ts)
q_gwpy = ts_gwpy.q_transform(frange=(30, 500))
fig = q_gwpy.plot()
fig.colorbar()
```

```python
ts_gwpy = TimeSeries.from_pycbc(ts)
q_gwpy = ts_gwpy.q_transform(frange=(30, 500), outseg=(60, 70))
fig = q_gwpy.plot()
fig.colorbar()
```
